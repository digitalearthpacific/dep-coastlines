```python
import rioxarray as rx
import odc.geo.xr
from rasterio.enums import Resampling
from dea_tools.spatial import subpixel_contours
```

```python
x = rx.open_rasterio('../dep_ls_coastlines/coastlines/0-6-0-52/062/030/1999_2023/dep_ls_coastlines_coastlines_062_030_1999_2023.tif')
xten = x.sel(band=12)
xten.odc.explore()
```

```python
xten10 = xten.rio.reproject(resolution=5, resampling=Resampling.nearest, dst_crs=xten.rio.crs)
```

```python
xten10_2 = xten.interp(x=xten10.x, y=xten10.y, method="linear")
xten10_2.odc.explore()
```

```python
import xrspatial as xs
import folium

xten30_focal = xs.focal.mean(xten)

xten30_focal_lines = subpixel_contours(xten30_focal)
m = xten30_focal_lines.explore(color="black")
xten_focal_lines = subpixel_contours(xten)
xten_focal_lines.explore(m=m)
#xten.odc.explore(map=m)
xten30_focal.odc.explore(map=m)
folium.LayerControl().add_to(m) 

m
```
